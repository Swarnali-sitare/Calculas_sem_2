GRADIENT DESCENT METHOD

```python
from sympy import *

x=symbols('x')

exp1=exp(x)-log(x)          # function taken

difn=diff(exp1,x)           # first derivative
dif1=difn.doit()

x_values=[]

tol=0.000001                # tolerance taken
step=0.2                    # step-size taken

# Initializing the values of xn_1 and xn
# where xn is the initial guess
xn_1=0
xn=1

iter=0                      # variable taken to count no. of iterations
while abs(xn-xn_1)>tol:
    
    iter+=1                             # increment of 1 each time the loop iterates
    xn_1=xn                             # changing the value of xn_1 each time the loop iterates
    subt=dif1.subs(x,xn_1)              # first derivative of the function
    
    xn=xn_1-(step*subt.evalf())         # formula used for Gradient Descent Method
    
    x_values.append(round(xn,10))       # appending values to list for graph plotting

print("minima at:",xn)

print("minimum value:",exp1.subs(x,xn))

print("total iterations:",iter)
```

NEWTON RAPHSON METHOD

```python
from sympy import *
x=symbols('x')

exp1=exp(x)-log(x)      # function taken

difn1=diff(exp1,x)      # first derivative
dif1=difn1.doit()       

difn2=diff(dif1,x)      # second derivative
dif2=difn2.doit()

X_values=[]

tol=0.000001            # tolerance taken

# initializing the values of xn_1 and xn
# where initial guess is xn
xn_1=0
xn=1

iter1=0                 # variable to count no. of iterations

while abs(xn-xn_1)>tol:
    
    xn_1=xn                                 # changing the value of xn_1 each time the loop iterates
    iter1+=1                                # increment of 1 each time the loop iterates
    subt1=dif1.subs(x,xn_1)                 # first derivative of the function on xn_1
    subt2=dif2.subs(x,xn_1)                 # second derivative of the function on xn_1
    
    xn=xn_1-(subt1.evalf()/subt2.evalf())   # formula used for Newton Raphson Method
    
    X_values.append(round(xn,10))           # appending values to list for graph plotting

print("minima at:",xn)

print("minimum value:",exp1.subs(x,xn))

print("total iterations:",iter1)
```

PLOTTING OF GRAPH (Iteration VS Values of xn)

```python
import matplotlib.pyplot as mp
import numpy as np

# for gradient descent method
pt_x=np.array(range(1,iter+1))
pt_y=np.array(x_values)

mp.plot(pt_x,pt_y,'o--k',ms=5,linewidth='0.5',label='Gradient Descent')
mp.xticks(pt_x,pt_x)

for i in range(len(pt_x)):
    mp.annotate(f'({pt_y[i]})', (pt_x[i], pt_y[i]), 
    fontsize=7, color='purple', fontweight='bold', 
    horizontalalignment='left', verticalalignment='bottom', 
    xytext=(0,3), textcoords='offset points')

# for newton raphson method
pt_X=np.array(range(1,iter1+1))
pt_Y=np.array(X_values)

mp.plot(pt_X,pt_Y,'x--r',ms=5,linewidth='0.5',label='Newton Raphson')

for i in range(len(pt_X)):
    mp.annotate(f'({pt_Y[i]})', (pt_X[i], pt_Y[i]), 
    fontsize=7, color='blue', fontweight='bold', 
    horizontalalignment='center', verticalalignment='top', 
    xytext=(0,-5), textcoords='offset points')

mp.legend()

mp.show()
```
